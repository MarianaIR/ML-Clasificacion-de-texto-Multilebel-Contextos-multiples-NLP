### **EXPLORANDO SOLUCIONES PARA USAR STRING COMO TARGET**

Para mejorar esto tenemos que numerizar los STRINGS, convertir el texto en numero, ayuda a que el calculo del modelo sea mas rapido sino que la precision del modelo sea mucho mejor que si estuviera trabajando con String.

```python
# 1°) Dividimos las etiquetas en etiquetas individuales, creamos un FOR --->

lista_de_tags = list() # Lista vacia
for tags in preguntas.Tags.unique(): # Para cada Tags recorre cada etiqueta en Preguntas.
  for tag in tags.split(): # Para cada uno de los tag, de los 8 si hay alguno compuesto dividelo
    if tag not in lista_de_tags: # Si esa eqtiqueta NO se encuentra en mi lista de etiquetas,
      lista_de_tags.append(tag) # La adicionamos

print(lista_de_tags)

# Nos devuelve las etiquetas(Tags) indivuduales.
```

### **TRANSFORMANDO TAGS EN COLUMNAS**

Ya tenemos las Etiquetas separadas individualmente, ahora tenemos que crear las columnas de cada etiqueta.

```python
# 1) Vamos a crear las columnas al DataFrame y vamos a adicionar 0 o 1 si la etiqueta de la columna Pregunta corresponde o NO a la
# columna de las etiquetas, para esto creamos un For --->

lista = list() # Lista vacia
for fila_tag in preguntas.Tags: # Recorremos la fila de Tags, la columna donde estan las etiquetas.
  if "list" in fila_tag: # Si se encuentra la palabra LIST en Tags
    lista.append(1) # Adicionamos a la lista vacia en forma de columna con el valor 1 si se encuentra esa etiqueta
  else:
    lista.append(0) # S NO la encuentra adiciona 0
preguntas["list"] = lista # Para que adicione al dataframe la columna list con el valor


preguntas.sample(15)

# Estamos transformando nuemericamente las etiquetas.
```

```python
# 2°) Para NO tener que crear una por una las etiquetas que NO es la forma correcta de hacerlo ya que en este caso son 5 pero podriamos
# tener 100 etqiuetas, creamos una FUNCION para crear columnas automaticamente con sus valores --->

def nueva_columna(lista_de_tags, dataframe, nombre_tags): # etiquetas, dataframe, nombre donde se encuentra la etiquetas.
  for tag in lista_de_tags: # Para cada una de las etiquetas que se encuentra en lista Tags
    columna = list() # Lista vacia
    for fila_tag in dataframe[nombre_tags]: # Para cada una de las filas en la columna Tags vamos a preguntar,
      if tag in fila_tag: # La etiqueta(5) se encuentra en Fila_tag del dataframe, si se encuentra,
        columna.append(1) # coloca valor 1
      else:
        columna.append(0) # coloca 0
    dataframe[tag] = columna # Nombre de la columna igual al nombre de cada una de las etiquetas que se encuentr en tag y los valores
```

```python
# 3°) Llamamos a la Funcion --->
nueva_columna(lista_de_tags, preguntas, "Tags") # Etiquetas, Dataframe, Nombre de la columna donde las etiquetas

preguntas.sample(5)

# Transformamos las Etiquetas a numeros.
```

### **ZIP PARA LA CLASIFICACION MULTILABEL**

Debemos combinar las 5 columnas numericas en 1 sola.

```python
# EJEMPLO --->
lista_1 = "1"
lista_2 = "2"
lista_zip = zip(lista_1, lista_2) # Metodo ZIP nos ayuda a juntar(zipar) elementos y nos devuelve un obeto ZIP, objeto Iterable.

print(lista_zip)

# Nos genero un objeto ZIP, para ver lo que contiene esto debemos convertir esto en una lista.
```

```python
# Ahora queremos ver que tenemos dentro de ese objeto y para esto le agregamos LIST:
print(list(lista_zip))

# Juntamos 2 objetos, uno con el valor 1 y otro con el valor 2 en un unico elemento.
```

```python
# 1°) Ahora juntamos las 5 columnas en una sola columna con el metodo ZIP, primero vemos los tags donde los guardamos anteriormente --->

print(lista_de_tags)
```

```python
# Ahora hacemos la lista zip --->

lista_zip_tags = list(
    zip(preguntas[lista_de_tags[0]], # preguntas=dataframe, [0]=list
        preguntas[lista_de_tags[1]], # [1]=arrays
        preguntas[lista_de_tags[2]],
        preguntas[lista_de_tags[3]],
        preguntas[lista_de_tags[4]]
        )
)
# Creamos nueva COLUMNA con todas las TAGS con el metodo de ZIP:
preguntas["todas_tags"] = lista_zip_tags
```

```python
# 2°) Ahora vemos las 5 primeras muestras con sample --->

preguntas.sample(5)

# Nos devuelve la nueva columna con los valores de todas las etiquetas con los valores respectivos.
```

### **TD-IDF**

Ya tenemos 1 unica columna con la Variable Predectiva(Variable Dependiente) y ahora podemos dividir los datos.

```python
# 1°) Dividimos los datos en ENTRENAMIENTO y TEST --->

preguntas_train, preguntas_test, tags_train, tags_test = train_test_split(
    preguntas.Preguntas, # Columnas Preguntas
    preguntas.todas_tags, # Columna de todas las tags
    test_size = 0.2, # 80% para Train, 20% Test
    random_state = 123
)
```

<font color=orage>***Vamos a utilizar una biblioteca llamada TF-IDF es una biblioteca que se utiliza mucho para NLP, nos ayuda a transformar el lenguaje humano a maquina, lo que hace es una FRECUENCIA de palabras, calcula la cantidad de veces que una palabra se repite en el texto y coloca una valor segun la cantidad de veces que se repita, si la palabra es insignificante le da un valor bao ya que se repite siempre, NO tiene mucho peso, menos frecuente se vuelve mas significante, mayor peso, y trabaa con estos valores altos que no son tan frecuentes para que trabaja la maquina***</font>

**¿cuál representa correctamente la principal característica del TF-IDF?**

* La principal característica del TF-IDF es que es una puntuación proporcional a la frecuencia de la palabra en el texto y equilibrada por la frecuencia en el corpus, es decir, las palabras que se repiten mucho suelen tener puntuaciones más bajas y son menos relevantes en el proceso de clasificación.
Estas son las principales características del TF-IDF.

### **COMPRENDIENDO ML-KNN**

En esta última clase vamos a conocer un modelo un poco más complejo que los anteriores. ¿Por qué más complejo? Porque en los anteriores dos modelos que hemos conocido nosotros tuvimos que transformar nuestra clasificación multilabel a una clasificación binaria, donde vamos pronosticando etiqueta por etiqueta individualmente.

Este tercer modelo ya utiliza el valor completo, el valor total de los múltiples valores como un único valor como variable predictora. ¿Qué modelo realiza este cálculo tan complejo para nosotros? El modelo ML-KNN.

El modelo ML-KNN es una variación del modelo super hiper conocido en el mundo de Machine Learning KNN, que es el de vecinos más próximos. Es un modelo que se utiliza para clasificación supervisionada. El modelo ML-KNN es una variación exclusiva para multilabel.
Multilabel KNN, por eso el nombre. Ahora vamos a intentar explicar cómo funciona primero el modelo KNN.

**¿Cómo funciona KNN?**

Bueno, imagínese. El modelo KNN imagínense que esto es un gráfico X donde tengo mi eje X y donde tengo mi eje y, aquí verticalmente. El modelo KNN representa matemáticamente todos tus datos dentro de este cuadrante y los va colocando según la distancia entre cada uno de mis datos. Por ejemplo, yo tengo 22700 preguntas en nuestro DataFrame, entonces este modelo KNN va a construir una representación matemática y decidir, según la distancia euclidiana dónde se encuentran cada una de mis 22700 preguntas y los va a ir distribuyendo en todo el gráfico.

Este es un ejemplo de menos preguntas que ya fueron distribuidas todas las preguntas de nuestro DataFrame y fueron colocadas dentro de nuestra representación gráfica según la distancia entre las etiquetas, entre los valores de las etiquetas de todas mis preguntas.
En este lado de aquí, yo tengo bastantes preguntas de pandas que están con una distancia más corta unas de las otras. Por este otro lado, tengo mis preguntas de Numpy que están con una distancia cortas entre ellas porque son preguntas relacionadas de Numpy. Por acá están mis preguntas relacionadas con pandas.

Y las de amarillo son las preguntas nuevas, son valores nuevos. Es decir, valores que no han sido todavía localizados en mi gráfico y el modelo va a intentar decidir a qué etiqueta corresponde, calculando la distancia de estas preguntas que no conocemos con los valores que ya se conocen y que ya se encuentran posicionados más cerca de esta pregunta.

Por ejemplo, vamos a trabajar primero con la pregunta 1. El modelo de KNN nos va a predecir cuántos vecinos más próximos necesitamos encontrar alrededor de esta pregunta. Nosotros vamos a buscar tres vecinos más próximos.
Entonces el modelo va a diseñar un círculo utilizando esa pregunta 1 como su eje principal, como su centro, y va a diseñar un círculo y va a buscar los tres vecinos más próximos, o sea, los que están más cercanos a mi dato, a mi pregunta de test, al dato que yo no conozco iba a encontrar que solamente tiene vecinos pandas.
Entonces el modelo va a decir: los tres vecinos más próximos que tienes a esta pregunta que no sé de qué se trata, son preguntas de pandas, así que yo te voy a decir que esta pregunta 1 es de pandas también. Y ya está. Ese es el pronóstico que realiza el modelo de KNN de vecinos más próximos.

Vamos a la pregunta número 2, aquí tenemos la pregunta número 2. También va a buscar los tres vecinos más próximos. Entonces va a comenzar a diseñar un círculo que se va a ir expandiendo hasta encontrar los tres vecinos más próximos. En este ejemplo, el modelo también nos va a decir: en esta pregunta número 2, yo he encontrado tres vecinos más próximos y todos son NumPy, así que yo voy a indicar que esta pregunta 2 también es de Numpy.

Es así cómo funciona KNN. Ahora tenemos la pregunta 3. La pregunta 3 es lo mismo. El modelo va a comenzar a crear un círculo alrededor de la pregunta 3, hasta encontrar los tres vecinos más próximos. Y va a crear un ciclo mayor, porque los vecinos más próximos están un poco más lejos en este caso.

Pero terminó de construir el círculo, encontró los tres vecinos más próximos. Están viendo que comienzan a aparecer etiquetas ya alrededor. Y aquí el modelo de KNN va a decir: tengo dos de Pandas y tengo una de Numpy. Así que según mi cálculo estadístico, tienes más chances de encontrar etiquetas de Pandas que de Numpy, así que yo te voy a decir que esta es una pregunta de Pandas.
Es así cómo funciona el modelo de KNN. Coge la etiqueta más frecuente dentro de mis tres vecinos más próximos.

**¿Cómo funciona el modelo ML-KNN?**

El modelo ML-KNN va a volver a insistir al modelo KNN que ya escogió las etiquetas para mi pregunta y le va a decir: “¿Estás completamente seguro que no has encontrado ningún otro valor? ¿Puedes confirmar al 100% de que solamente vamos a encontrar pandas aquí?”

Y el modelo KNN comienza a conferir sus cálculos, y en la pregunta 1 me va a decir: “Sí. Estoy completamente seguro que solamente hay etiquetas de pandas, así que continúa siendo pandas. En la pregunta 2, sí estoy completamente seguro que hay etiquetas de Numpy, así que continúa siendo Numpy.”

Y en la pregunta 3. “En la pregunta 3, sí no estoy tan seguro, mira, te voy a ser sincero, he encontrado palabras de Numphy con una frecuencia alta, es decir, he encontrado en una proporción de una a dos preguntas de Numphy, así que el modelo KNN va a asumir también el valor de Numpy como una posible etiqueta y nos va a devolver como resultado que para la pregunta 3 tengo una clasificación multilabel, es decir, estoy encontrando dos etiquetas como posible resultado”.

Esto es, la manera como funciona el modelo de ML-KNN, vecinos más próximos para una clasificación multilabel
